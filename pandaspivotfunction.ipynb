```python
#pandas pivot table
#the levels in the pivot table will be stored in multiindex object(hierarchical indexes)
#on the index and columns of the result DataFrame.
#syntax: df.pivot_table()
#parameters:
#1. value 2.fill_value
#3.index 4.margin
#5.columns 6.dropna
#7.aggfunc
```

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\kaushal prasad\\Downloads\\dataset\\Movie_ratings.csv')
df
```

```python
df.pivot_table(index = 'Genre')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',aggfunc = 'count')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',aggfunc = 'max')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',aggfunc = 'sum')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',aggfunc = 'sum',fill_value = 0)
```

```python
#dropna True(Default)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',fill_value = 0, margins = True)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year',aggfunc = 'sum',fill_value = 0, margins = True)
```
