```python
#pandas melt function
#it is use to transform or reshape data.
#syntax: pd.melt
#parameter:
# 1.frame  2.var_name
# 3.id_vars 4.value_name
# 5.value_vars
```

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\kaushal prasad\\Downloads\\dataset\\Movie_ratings.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars = ['Genre'])
```

```python
pd.melt(df, id_vars = ['Year'])
```

```python
#unpivot
#a dataframe turn it from a wide format(many columns) to a long format(few columns but many rows).
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Genre'])
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Ratings %'])
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Ratings %'], var_name = 'Category')
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Ratings %'], var_name = 'Category', value_name = 'Data')
```
